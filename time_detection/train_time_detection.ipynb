```python
# Path to the directory containing the images and text files
directory = '/content/drive/MyDrive/VisionSEO_Project/time_detection/data'

# Get a list of all image and text files in the directory
image_files = {f.split('.')[0] for f in os.listdir(directory) if f.endswith(('.png', '.jpg'))}
text_files = {f.split('.')[0] for f in os.listdir(directory) if f.endswith('.txt')}

# Find the images that don't have a corresponding text file
images_to_remove = image_files - text_files

# Remove the images that don't have a corresponding text file
for img in images_to_remove:
    for ext in ['.png', '.jpg']:
        img_path = os.path.join(directory, img + ext)
        if os.path.exists(img_path):
            os.remove(img_path)
            print(f'Removed {img_path}')
```

```python
import yaml

# Define paths and classes
dataset_yaml_path = '/content/drive/MyDrive/VisionSEO_Project/time_detection/data/dataset.yaml'
images_train_path = '/content/drive/MyDrive/VisionSEO_Project/time_detection/data/images/train'
labels_train_path = '/content/drive/MyDrive/VisionSEO_Project/time_detectiondata/labels/train'

classes = [
    'dog', 'person', 'cat', 'tv', 'car', 'meatballs', 'marinara sauce',
    'tomato soup', 'chicken noodle soup', 'french onion soup', 'chicken breast',
    'ribs', 'pulled pork', 'hamburger', 'cavity', 'time_date'
]

# Prepare the YAML content
yaml_content = {
    'path': '/content/drive/MyDrive/VisionSEO_Project/time_detection/data',
    'train': {
        'images': images_train_path,
        'labels': labels_train_path
    },
    'val': {
        'images': images_train_path,  # Assuming you use the same for validation
        'labels': labels_train_path   # Assuming you use the same for validation
    },
    'nc': len(classes),
    'names': classes
}

# Write to dataset.yaml
with open(dataset_yaml_path, 'w') as file:
    yaml.dump(yaml_content, file, sort_keys=False)

print(f"Dataset YAML file created at: {dataset_yaml_path}")
```

```python
# Convert results to a format where we can extract bounding boxes
boxes = results[0].boxes

# Check if there are any detections
if len(boxes) == 0:
    print("No detections found in the image.")
else:
    # Load the original image
    image = Image.open('/content/drive/MyDrive/VisionSEO_Project/screenshots/Screenshot 2024-08-08 at 13.53.36 - Sonal Edwin.png')

    # Loop over the detections
    for box in boxes:
        cls_id = int(box.cls)  # Class ID
        cls_name = results[0].names[cls_id]  # Get class name

        # Get bounding box coordinates
        x1, y1, x2, y2 = map(int, box.xyxy[0].numpy())

        # Crop the detected region
        cropped_img = image.crop((x1, y1, x2, y2))

        # Convert the cropped image to OpenCV format
        cropped_img_cv = cv2.cvtColor(np.array(cropped_img), cv2.COLOR_RGB2BGR)

        # Apply OCR using Tesseract
        text = pytesseract.image_to_string(cropped_img_cv)

        # Print the extracted text
        print(f"Detected {cls_name}: {text.strip()}")

        # Optionally, display the cropped region and the extracted text
        plt.figure(figsize=(5, 5))
        plt.imshow(cropped_img)
        plt.title(f"{cls_name} - Extracted Text: {text.strip()}")
        plt.axis('off')
        plt.show()
```
